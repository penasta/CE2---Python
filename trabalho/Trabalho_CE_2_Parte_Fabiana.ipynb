Pacotes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Importar dados

```python
costumers = pd.read_csv('https://raw.githubusercontent.com/edunb01/dotfiles/master/olist_customers_dataset.csv')
itens = pd.read_csv('https://raw.githubusercontent.com/edunb01/dotfiles/master/olist_order_items_dataset.csv')
payments = pd.read_csv('https://raw.githubusercontent.com/edunb01/dotfiles/master/olist_order_payments_dataset.csv')
reviews = pd.read_csv('https://raw.githubusercontent.com/edunb01/dotfiles/master/olist_order_reviews_dataset.csv')
orders = pd.read_csv('https://raw.githubusercontent.com/edunb01/dotfiles/master/olist_orders_dataset.csv')
products = pd.read_csv('https://raw.githubusercontent.com/edunb01/dotfiles/master/olist_products_dataset.csv')
sellers = pd.read_csv('https://raw.githubusercontent.com/edunb01/dotfiles/master/olist_sellers_dataset.csv')
```

Unir dados em um único bando de dados

```python
df = itens.merge(payments)
df = df.merge(sellers)
df = df.merge(products)
df = df.merge(reviews)
df = df.merge(orders)
df = df.merge(costumers)
df
```

Vendedores por Estados (para comparações futuras)

```python
# Criar um gráfico de barras para mostrar a distribuição dos vendedores por estado
cont_vendedores_por_estado = df['seller_state'].value_counts()
plt.figure(figsize=(20, 10))
vendedores_estado = cont_vendedores_por_estado.plot(kind='bar', color='gold')
plt.xlabel('Estado')
vendedores_estado.set_xticklabels(vendedores_estado.get_xticklabels(), rotation=0)
plt.ylabel('Número de Vendedores')
plt.yticks(range(0, cont_vendedores_por_estado.max() + 1, 2500))
plt.title('Distribuição de Vendedores por Estado')
plt.show()
```

Vendedores com produtos mais caros

```python
# Ordenar o DataFrame pelo preço em ordem decrescente
df_preco = df.sort_values(by='price', ascending=False)

# Agrupar por preço
df_top_precos = df_preco.groupby('price').first().reset_index()

# Contar o número de vendedores
num_vendedores = df_top_precos['seller_id'].nunique()

# Exibir o número de vendedores
print(num_vendedores)
```

```python
df_top_precos
```

```python
# Obter medidas de posição do DataFrame para os preços
medidas_posicao = df_top_precos['price'].describe()

# Obter os valores dos quantis 90%, 93%, 94%, 95%, 98% e 99% para os preços (mais altos)
quantis = df_top_precos['price'].quantile([0.90, 0.93, 0.94, 0.95, 0.98, 0.99])

# Exibir medidas de posição e quantis
print(medidas_posicao)
print(quantis)
```

```python
# Selecionar 2% maiores valores
top_2_porcento = df_top_precos.nlargest(int(0.02 * len(df_top_precos)), 'price')

# Contar o número de vendedores nos 2% dos produtos mais caros
num_vendedores_2_porcento = top_2_porcento['seller_id'].nunique()

# Exibir o número de vendedores nos 2% dos produtos mais caros
print(num_vendedores_2_porcento)
```

```python
# Criar um gráfico de barras horizontais para os 2% maiores valores
plt.figure(figsize=(10, 15))
plt.barh(top_2_porcento['seller_id'], top_2_porcento['price'], color='pink')
plt.xlabel('Preço')
plt.ylabel('Vendedor')
plt.title('Top 2% dos Produtos Mais Caros por Vendedor')
plt.show()
```

```python
# Selecionar 1% maiores valores
top_1_porcento = df_top_precos.nlargest(int(0.01 * len(df_top_precos)), 'price')

# Contar o número de vendedores nos 1% dos produtos mais caros
num_vendedores_1_porcento = top_1_porcento['seller_id'].nunique()

# Exibir o número de vendedores nos 1% dos produtos mais caros
print(num_vendedores_1_porcento)
```

Valor do pagamento versus Número de parcelas

```python
# Obter as medidas de posição para as duas variáveis
medidas_posicao_valor = df['payment_value'].describe()
medidas_posicao_parcelas = df['payment_installments'].describe()

# Obter os valores dos quantis 90%, 95%, 98% e 99%
quantis_valor = df['payment_value'].quantile([0.90, 0.95, 0.98, 0.99])
quantis_parcelas = df['payment_installments'].quantile([0.90, 0.95, 0.98, 0.99])

# Exibir medidas de posição e os quantis
print(medidas_posicao_valor)
print(quantis_valor)
print(medidas_posicao_parcelas)
print(quantis_parcelas)
```

```python
# Criar um gráfico de dispersão com as variáveis invertidas
plt.figure(figsize=(20, 10))
plt.scatter(df['payment_value'], df['payment_installments'], color='darkred')
plt.xlabel('Valor do Pagamento')
plt.ylabel('Número de Parcelas')
plt.yticks(range(0, 25, 1))
plt.title('Valor do Pagamento vs Parcelas')
plt.show()
```

```python
# Calcular a correlação de Pearson entre as duas variáveis
correlacao = df['payment_value'].corr(df['payment_installments'])

print(f'A correlação entre as variáveis é: {correlacao}')
```

O valor da correlação de Pearson próximo de 0 indica uma correlação fraca ou inexistente. O gráfico permite visualizar que o número de parcelas não varia de acordo com o valor da compra. De fato, a maior parte das compras é paga entre 1 e 10 parcelas com valores de pagamento até, aproximadamente, 7.500,00 reais, com um valor de 13.664,00 reais outlier. Enquanto, as compras parceladas entre 11 e 24 vezes apresentam valores de pagamento até 2.000 reais, aproximadamente.

As medidas-resumo e os quantiscorroboram o gráfico, 75% das compras são pagas em até 4 parcelas e 99% em até 10 parcelas. Os valores encontrados para o quantis, indicam que o gráfico de pontos apresenta um falha para visualização, as ocorrências de pagamentos em até 10 parcelas ficam subrepresentadas, pois os pontos se sobrepõem.
