```python
from scipy import stats
import seaborn as sns
import pandas as pd
from scipy.stats import norm
import numpy as np
import statistics
```

```python
pop = stats.norm.rvs(loc=100,scale=5, size = 100000)
#pop
```

```python
n = 100
estimativas = []
for i in range(1000):
    amostra = pop[stats.randint.rvs(0,100000,size=n)]
    dp = amostra.var()
    estimativas.append(dp)
#estimativas
sns.histplot(estimativas,kde=True)
```

```python
np.array(estimativas).mean()
```
