## Questão 1

Considerando a distribuição de Poisson com média $\mu=5$  faça:

* Simule um conjunto de 20 valores aleatórios e calcule a frequência relativa dos valores observados

* Para os valores observados no item anterior calcule a probabilidade de ocorrência segundo a distribuição de Poisson $\mu=5$ e compare os resultados com as frequências calculadas no item anterior.

* Repita os itens anteriores mas desta vez gerando um conjunto de 2000 valores e comente sobre os resultados obtidos.


```python
import numpy as np
data = np.random.poisson(5, 20)
data
```

```python
import math
for i in range(0, len(np.unique(data))):
    prob = math.exp(-5)*math.pow(5, np.unique(data)[i])/math.factorial(np.unique(data)[i])
    print("Valor:",np.unique(data)[i],"Probabilidade:",prob)
```

```python
data2 = np.random.poisson(5, 2000)
for i in range(0, len(np.unique(data2))):
    prob = math.exp(-5)*math.pow(5, np.unique(data2)[i])/math.factorial(np.unique(data2)[i])
    print("Valor:",np.unique(data2)[i],"Probabilidade:",prob)
```
