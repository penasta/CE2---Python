```python
from scipy import stats
import numpy as np
import seaborn as sns
```

```python
xcrit = stats.norm.ppf(q = .05,loc=100,scale=(9/30)**(1/2))
```

## Erro tipo 1:

```python
resultado = []
for i in range (10000):
    amostra = stats.norm.rvs(loc=100,scale=3,size=30)
    resultado.append(amostra.mean() < xcrit)
np.array(resultado).mean()
```

## Erro tipo 2:

```python
resultado = []
for i in range (10000):
    amostra = stats.norm.rvs(loc=98,scale=3,size=30)
    resultado.append(amostra.mean() > xcrit)
beta = np.array(resultado).mean()

print("Probabilidade de erro tipo II:",beta,"Poder do teste:",1-beta)
```

```python
1 - stats.norm.cdf(xcrit,loc=[97.5,98,98.5,99],scale=(9/30)**(1/2))
```

```python
valores_alternativos=np.linspace(95,100,num=100)
#valores_alternativos
erro2 = 1 - stats.norm.cdf(xcrit,loc=valores_alternativos,scale=(9/30)**(1/2))
sns.lineplot(x=valores_alternativos,y=erro2)
```

```python
poder = 1- erro2
sns.lineplot(x=valores_alternativos,y=poder)
```
