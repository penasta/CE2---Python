```python
from scipy import stats
import numpy as np
import seaborn as sns
```

```python
populacao = stats.norm.rvs(loc=100,scale= 5, size=100000)
populacao
populacao.var()
```

```python
resultados = []
for i in range(10000):
    amostra = populacao[stats.randint.rvs(0,100000,size=100)]
    resultados.append(amostra.var())
sns.histplot(resultados)
```

```python
np.array(resultados).mean()
```

```python
resultados = []
amostra = populacao[stats.randint.rvs(0,100000,size=30)]
for i in range(1000):
    reamostra = amostra[stats.randint.rvs(0,30,size=30)]
    resultados.append(reamostra.var())

sns.histplot(resultados)
```

```python
amostra.var()
```

```python
np.quantile(resultados,[.05,.95])
```

```python
varpop = populacao.var()
resultados = []
for i in range(1000):
    variancias = []
    amostra = populacao[stats.randint.rvs(0,100000,size = 30)]
    for j in range(100):
        reamostra = amostra[stats.randint.rvs(0,30,size=30)]
        variancias.append(reamostra.var())
    resultados.append((np.quantile(variancias,.05)<varpop) and (np.quantile(variancias,.95)>varpop))
np.array(resultados).mean()
```
